```python
# Book prices
prices = [120, 225, 354, 413, 622, 175, 281]

# 1) Extract elements in range 200 to 300
def range_200_300(data):
    return [x for x in data if 200 <= x <= 300]

# 2) Display elements except 2nd through 5th
def exclude_2_to_5(data):
    return data[:1] + data[5:]

# 3) Display elements > 140 and < 470
def between_140_470(data):
    return [x for x in data if 140 < x < 470]

# 4) Display elements > 600
def greater_than_600(data):
    return [x for x in data if x > 600]

print(range_200_300(prices))
print(exclude_2_to_5(prices))
print(between_140_470(prices))
print(greater_than_600(prices))
```
