```python
import numpy as np

# Data
age = np.array([23, 23, 52, 54, 54, 56, 57, 58, 58, 60, 61])
fat = np.array([9.5, 26.5, 34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7])

# i) Mean
mean_age = np.mean(age)
mean_fat = np.mean(fat)

# Median
median_age = np.median(age)
median_fat = np.median(fat)

# Standard Deviation
std_age = np.std(age, ddof=1)
std_fat = np.std(fat, ddof=1)

print("Mean Age:", mean_age)
print("Mean %Fat:", mean_fat)

print("Median Age:", median_age)
print("Median %Fat:", median_fat)

print("Standard Deviation of Age:", std_age)
print("Standard Deviation of %Fat:", std_fat)

# ii) Pearson Correlation Coefficient
correlation = np.corrcoef(age, fat)[0, 1]
print("Pearson Correlation Coefficient:", correlation)

# Relationship type
if correlation > 0:
    print("Age and %Fat are positively correlated.")
elif correlation < 0:
    print("Age and %Fat are negatively correlated.")
else:
    print("No correlation between Age and %Fat.")
```
