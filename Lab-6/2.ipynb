```python

import numpy as np

YoE = np.array([1, 1.1, 1.3, 2, 2.2, 2.7, 3, 3.2, 3.9, 4, 4.1, 4.2])
SAL = np.array([32323, 45207, 39751, 43525, 39891, 56542,
                60150, 54545, 63218, 55794, 56081, 57081])

n = len(YoE)

x_mean = np.mean(YoE)
y_mean = np.mean(SAL)

numerator = np.sum((YoE - x_mean) * (SAL - y_mean))
denominator = np.sum((YoE - x_mean) ** 2)
beta_1 = numerator / denominator

beta_0 = y_mean - beta_1 * x_mean

SAL_pred = beta_0 + beta_1 * YoE

error = SAL - SAL_pred

print("Mean of YoE:", x_mean)
print("Mean of SAL:", y_mean)

print("\nSlope (Beta1):", beta_1)
print("Intercept (Beta0):", beta_0)

print("\nPredicted Salaries:")
print(SAL_pred)

print("\nRandom Errors (Residuals):")
print(error)

if beta_1 > 0:
    relation = "Positive Linear Relationship"
elif beta_1 < 0:
    relation = "Negative Linear Relationship"
else:
    relation = "No Linear Relationship"

print("\nRelationship between YoE and SAL:", relation)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(YoE, SAL, label="Actual Data")
plt.plot(YoE, SAL_pred, label="Regression Line")
plt.xlabel("Years of Experience (YoE)")
plt.ylabel("Salary (SAL)")
plt.title("Linear Regression: YoE vs Salary")
plt.legend()
plt.grid(True)
plt.show()
```
