```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
age = np.array([43,21,25,42,57,59]).reshape(-1,1)
glucose = np.array([99,65,79,75,87,81])

# Model
model = LinearRegression()
model.fit(age, glucose)

# Plot
plt.scatter(age, glucose)
plt.plot(age, model.predict(age))
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.title("Age vs Glucose Level")
plt.show()

print("Equation: Glucose =", model.coef_[0], "* Age +", model.intercept_)
```
