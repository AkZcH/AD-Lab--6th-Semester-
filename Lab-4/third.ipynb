```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
competitors = np.array([1,1,2,3,3,5,5,6]).reshape(-1,1)
sales = np.array([3600,3300,3100,2900,2700,2300,2000,1800])

# Model
model = LinearRegression()
model.fit(competitors, sales)

# Prediction
x_line = np.linspace(1,6,100).reshape(-1,1)
y_line = model.predict(x_line)

# Plot
plt.scatter(competitors, sales)
plt.plot(x_line, y_line)
plt.xlabel("Number of Competitors")
plt.ylabel("Sales Volume")
plt.title("Sales vs Competitors")
plt.show()

print("Regression Equation: Sales =", model.coef_[0], "* Competitors +", model.intercept_)
```
