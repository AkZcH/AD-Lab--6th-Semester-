```python
import numpy as np
from sklearn.linear_model import LogisticRegression

hours = np.array([1, 2, 3, 5]).reshape(-1, 1)
probability = np.array([0.07, 0.26, 0.61, 0.97])

y = (probability >= 0.5).astype(int)

model = LogisticRegression()
model.fit(hours, y)

new_hours = np.array([[4]])
predicted_prob = model.predict_proba(new_hours)[0][1]
prob_percent = predicted_prob*100

print("Predicted probability of passing for 4 hours of study:", predicted_prob)
print("Probability Percentage: ", {prob_percent}, "%")
```
