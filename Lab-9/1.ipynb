```python
import numpy as np

# ---------------- BREAM ----------------
length_bream = np.array([25.4, 26.3, 26.5, 29, 29.7, 37])
weight_bream = np.array([242, 290, 340, 363, 500, 1000])

n = len(length_bream)

b_bream = (n*np.sum(length_bream*weight_bream) - np.sum(length_bream)*np.sum(weight_bream)) / \
          (n*np.sum(length_bream**2) - (np.sum(length_bream))**2)

a_bream = np.mean(weight_bream) - b_bream*np.mean(length_bream)

print("Bream Model:")
print(f"Weight = {a_bream:.2f} + {b_bream:.2f} * Length")


# ---------------- ROACH ----------------
length_roach = np.array([23.5, 25.2, 26, 31.7, 22.5, 20.8, 14.5])
weight_roach = np.array([200, 180, 290, 390, 160, 140, 40])

n = len(length_roach)

b_roach = (n*np.sum(length_roach*weight_roach) - np.sum(length_roach)*np.sum(weight_roach)) / \
          (n*np.sum(length_roach**2) - (np.sum(length_roach))**2)

a_roach = np.mean(weight_roach) - b_roach*np.mean(length_roach)

print("\nRoach Model:")
print(f"Weight = {a_roach:.2f} + {b_roach:.2f} * Length")
```
