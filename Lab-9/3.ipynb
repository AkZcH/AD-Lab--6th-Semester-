```python
import numpy as np

# Class intervals midpoints
midpoints = np.array([4.5, 14.5, 24.5, 34.5, 44.5, 54.5, 64.5, 74.5, 84.5])
frequency = np.array([20, 21, 23, 16, 11, 10, 7, 3, 1])

N = frequency.sum()

# -------- MEAN --------
grouped_mean = np.sum(midpoints * frequency) / N
print("Estimated Mean:", grouped_mean)


# -------- MEDIAN --------
cum_freq = np.cumsum(frequency)

median_class_index = np.where(cum_freq >= N/2)[0][0]

L = 19.5        # lower boundary of median class
cf_before = cum_freq[median_class_index - 1]
f = frequency[median_class_index]
w = 10

median = L + ((N/2 - cf_before) / f) * w
print("Estimated Median:", median)


# -------- MODE --------
fm = frequency[median_class_index]
f_prev = frequency[median_class_index - 1]
f_next = frequency[median_class_index + 1]

mode = L + ((fm - f_prev) / ((fm - f_prev) + (fm - f_next))) * w
print("Estimated Mode:", mode)
```
