```python
import pandas as pd
import math

data = pd.DataFrame([
["YELLOW","SMALL","DIP","ADULT","F"],
["YELLOW","LARGE","STRETCH","ADULT","T"],
["YELLOW","LARGE","STRETCH","CHILD","F"],
["YELLOW","LARGE","DIP","ADULT","F"],
["YELLOW","LARGE","DIP","CHILD","F"],
["PURPLE","SMALL","STRETCH","ADULT","T"],
["PURPLE","SMALL","STRETCH","ADULT","T"],
["PURPLE","SMALL","STRETCH","CHILD","F"],
["PURPLE","SMALL","DIP","ADULT","F"],
["PURPLE","SMALL","DIP","CHILD","F"]
], columns=["Color","Size","Act","Age","Inflated"])

def entropy(col):
    probs = col.value_counts(normalize=True)
    return -sum(p*math.log2(p) for p in probs)

target_entropy = entropy(data["Inflated"])

def info_gain(attr):
    total = len(data)
    weighted = sum(
        (len(sub)/total)*entropy(sub["Inflated"])
        for _, sub in data.groupby(attr)
    )
    return target_entropy - weighted

for col in ["Color","Size","Act","Age"]:
    print(col, "IG =", info_gain(col))
```
