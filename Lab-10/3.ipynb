```python
import pandas as pd

data = pd.DataFrame([
["Fast","No","No","Fish"],
["Fast","No","Yes","Animal"],
["Slow","No","No","Animal"],
["Fast","No","No","Animal"],
["No","Short","No","Bird"],
["No","Short","No","Bird"],
["No","Rarely","No","Animal"],
["Slow","No","Yes","Animal"],
["Slow","No","No","Fish"],
["Slow","No","Yes","Fish"],
["No","Long","No","Bird"],
["Fast","No","No","Bird"]
], columns=["Swim","Fly","Crawl","Class"])

X_new = ["Slow","Rarely","No"]

classes = data["Class"].unique()
posterior = {}

for c in classes:
    subset = data[data["Class"] == c]
    prob = len(subset)/len(data)

    for col, val in zip(["Swim","Fly","Crawl"], X_new):
        prob *= len(subset[subset[col]==val]) / len(subset)

    posterior[c] = prob

print("Prediction:", max(posterior, key=posterior.get))
```
