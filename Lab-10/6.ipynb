```python
import numpy as np
import pandas as pd
from math import sqrt, pi, exp

data = pd.DataFrame({
"X1":[5,7,6,8,5,9,6,7],
"X2":[12,15,14,18,10,20,13,17],
"Y":["A","B","A","B","A","B","A","B"]
})

def gaussian(x, mean, std):
    return (1/(sqrt(2*pi)*std)) * exp(-((x-mean)**2)/(2*std**2))

X_new = [6,16]
classes = data["Y"].unique()

posterior = {}

for c in classes:
    subset = data[data["Y"]==c]
    prior = len(subset)/len(data)

    p1 = gaussian(X_new[0], subset["X1"].mean(), subset["X1"].std())
    p2 = gaussian(X_new[1], subset["X2"].mean(), subset["X2"].std())

    posterior[c] = prior * p1 * p2

print("Prediction:", max(posterior, key=posterior.get))
```
