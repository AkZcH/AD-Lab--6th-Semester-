```python
TP = 22
FP = 10
FN = 8
TN = 60

accuracy = (TP + TN) / (TP + TN + FP + FN)
recall = TP / (TP + FN)
precision = TP / (TP + FP)
f1 = 2 * (precision * recall) / (precision + recall)

print("Accuracy :", accuracy)
print("Recall   :", recall)
print("Precision:", precision)
print("F1 Score :", f1)
```
