```python
import numpy as np
data = np.array([5,10,11,13,15,35,50,55,72,92,204,215])

# Equal frequency
eq_freq = np.array_split(data, 3)
print("Equal Frequency:", eq_freq)

# Equal width
bins = np.linspace(data.min(), data.max(), 4)
eq_width = np.digitize(data, bins)

print("Equal Width bins:", eq_width)
```
