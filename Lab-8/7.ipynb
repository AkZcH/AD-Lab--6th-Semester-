```python
import pandas as pd
import numpy as np
from math import log2

data = pd.DataFrame({
    "Temperature": ["High","High","Medium","Medium","High","Medium","High","Medium","Medium","High"],
    "Wind": ["South","West","West","South","West","South","West","South","South","South"],
    "Rainy": ["N","Y","N","N","N","Y","N","N","Y","N"],
    "Humidity": ["High","High","High","High","Medium","High","High","High","High","Medium"],
    "GoingOut": ["Y","Y","Y","N","N","N","Y","Y","N","Y"]
})

def entropy(col):
    probs = col.value_counts(normalize=True)
    return -sum(p * log2(p) for p in probs)

def info_gain(df, feature, target="GoingOut"):
    total_entropy = entropy(df[target])
    vals = df[feature].unique()
    weighted_entropy = sum(
        (len(df[df[feature]==v]) / len(df)) * entropy(df[df[feature]==v][target])
        for v in vals
    )
    return total_entropy - weighted_entropy

for col in data.columns[:-1]:
    print(col, info_gain(data, col))
```
