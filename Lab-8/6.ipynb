```python
import numpy as np
from collections import Counter

X = np.array([
    [4,35],
    [6,40],
    [3,25],
    [7,45],
    [5,30],
    [8,50],
    [2,20],
    [5,35]
])

y = np.array([
    "Cat","Dog","Cat","Dog","Cat","Dog","Cat","Dog"
])

new_point = np.array([4,30])

distances = np.linalg.norm(X - new_point, axis=1)
k = 5
nearest = distances.argsort()[:k]

prediction = Counter(y[nearest]).most_common(1)[0][0]
print("Predicted Species:", prediction)
```
